Implementation of chatbot

```python
import nltk
import random
import os
import ssl
import streamlit as st
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

```

```python
ssl._create_default_https_context = ssl._create_unverified_context
nltk.data.path.append(os.path.abspath('nltk_data'))
nltk.download('punkt')
```

```python
intents = [
    
 { 'tag': 'inquiry',
 'patterns': ['How are you?' , 'Can you explain this to me?'],
 'responses': ['I am doing great,how about you?','I would be happy to explain. What part would you like to know more about?']
 },

 {'tag': 'confirmation',
 'patterns': ['What is  your plan for today?', 'Is this correct?'],
 'responses': ['Yes, I did! Got it loud and clear.', 'Yes, that is correct.']
 },

 {'tag': 'apology',
 'patterns': ['Sorry for the delay,I got caught up.' ,'Apolozies for the confusion.'],
 'responses': ['No problem at all, I understand.', 'Do not worry,I understand!']
 },

 {'tag': 'request',
 'patterns': ['Can you send me that file?' , 'Would you mind helping me out?'],
 'responses': ['Sure,I will send it right away!', 'I would be glad to help.']
 },

 {'tag': 'gratitude',
 'patterns': ['I can not thank you enough','I really appreciate it.'],
 'responses': ['Most welcome, happy to help!', 'It was my pleasure! Let me know if you need anything else.']
 },

 {'tag': 'encouragement',
 'patterns': ['You can do it, do not give up!', 'I do not think I can do it.'],
 'responses': ['Thanks, I will keep going for sure!', 'It is okay to feel this! Keep going and you will see the progress soon.']
 },

 {'tag': 'surprise',
 'patterns': ['Wow, that is amazing!', 'Really? I am shocked!'],
 'responses': ['Right? I was shocked too!', 'It is always nice to be surprised.']
 },

 {'tag': 'farewell',
 'patterns': ['Goodbye' , 'Take care!'],
 'responses': ['See you later','Take care and stay safe.']
 },

 {'tag': 'weather',
 'patterns': ['Is it raining?', 'What is the weather like today?'],
 'responses': ['No, It is clear today', 'Let me quickly check the weather for you.']
 },

 {'tag': 'compliment',
 'patterns': ['You look great', 'You have such a kind personality.'],
 'responses': ['Thanks,so do you!', 'That is so kind of you to say!']
  }
]
```

```python
# Create the vectorizer and classifier
vectorizer = TfidfVectorizer()
clf = LogisticRegression(random_state=0, max_iter=10000)

#Preprocess the data
tags = []
patterns = []
for intent in intents:
    for pattern in intent['patterns']:
        tags.append(intent['tag'])
        patterns.append(pattern)
```

```python
#Training the model
x = vectorizer.fit_transform(patterns)
y = tags
clf.fit(x, y)
```

```python
def chatbot(input_text):
    input_text = vectorizer.transform([input_text])
    tag = clf.predict(input_text)[0]
    for intent in intents:
        if intent['tag'] == tag:
            response = random.choice(intent['responses'])
            return response
```

```python
#Checking our chatbot
user_input = "You look great"
response = chatbot(user_input)
print(response)
```

```python
user_input = "Is it raining?"
response = chatbot(user_input)
print(response)
```
